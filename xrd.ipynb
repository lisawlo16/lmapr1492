# Imports & fichier cif

```python
#Imports
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from matplotlib import pyplot as plt
import numpy as np

#téléchargement du fichier cif 
from pymatgen.ext.matproj import MPRester
with MPRester("hJ9SDZIQUxgk5cVi") as m:
    # Structure for material id
    struct = m.get_structure_by_material_id("mp-1367")
```

# Diffractogramme de la structure

```python
calc = XRDCalculator(wavelength='CuKa')
difpat = calc.get_pattern(struct, scaled = True)
calc.get_plot(struct, two_theta_range=(0,90))
plt.show()
```

# Premier pic

```python
print("Intensité = ", difpat.y[0],"%")
print("Angle 2\u0398 = ", difpat.x[0],"°")
print("Indices hkl = ", difpat.hkls[0], '\n')
```

# Deuxième pic

```python
print("Intensité = ", difpat.y[1],"%")
print("Angle 2\u0398 = ", difpat.x[1],"°")
print("Indices hkl = ", difpat.hkls[1],'\n')
```

# Troisième pic

```python
print("Intensité = ", difpat.y[2],"%")
print("Angle 2\u0398 = ", difpat.x[2],"°")
print("Indices hkl = ", difpat.hkls[2],'\n')
```

# Diffractogramme des 3 premiers pics d'absorption

```python
xaxis = [(difpat.x)[0],(difpat.x)[1],(difpat.x)[2]] 
yaxis = [(difpat.y)[0],(difpat.y)[1],(difpat.y)[2]] 
width = 0.3
plt.bar(xaxis, yaxis, width) 
plt.xlabel('2\u03B8')
plt.ylabel('Intensité')
plt.title('Diffractogramme des trois premiers pics')
plt.show()
```

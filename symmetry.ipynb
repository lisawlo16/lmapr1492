# Import et téléchargement fichier cif

```python
#import
from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from jupyter_jsmol.pymatgen import quick_view
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.operations import SymmOp
import matplotlib.pyplot as plt
import matplotlib.image as img

#téléchargement du fichier cif de mp-1367
with MPRester("hJ9SDZIQUxgk5cVi") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-1367")
```

# Positions et choix des symétries

```python
# Pour analyser la structure de notre molécule
space = SpacegroupAnalyzer(structure)

# Structure conventionnelle
struc_conv = space.get_conventional_standard_structure()

#Visualisation des positions des différents atomes
sites = struc_conv.sites
#view = quick_view(struc_conv)
#display(view)

# Liste des symétries
sym = SpacegroupAnalyzer(struc_conv).get_symmetry_operations()
```

# Opération de symétrie 1 : Inversion par rapport au point (0,0,0)

```python
print("Matrices de rotation et de translation de l'opération de symétrie :", "\n", sym[1], "\n")
position = sites[0].frac_coords
print("J'applique une inversion à l'atome de Mg de coordonnées :", position, "\n" )
symmetry = sym[1].operate(position)
print("Les coordonées de l'atome deviennent alors: ", symmetry)

# Affichage de l'image de la symétrie
image_1=img.imread('Inversion.jpg')
fig, axes = plt.subplots(figsize=(12,12))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

# Opération de symétrie 2 : Plan miroir défini par (-y,-x,z)

```python
print("Matrices de rotation et de translation de l'opération de symétrie :", "\n", sym[15], "\n")
position = sites[0].frac_coords
print("J'applique un plan miroir défini par (-y,-x,z) à l'atome de Mg de coordonnées :", position, "\n" )
symmetry = sym[15].operate(position)
print("Les coordonées de l'atome deviennent alors: ", symmetry)

#Affichage de l'image de la symétrie
image_2=img.imread('Plan.jpg')
fig, axes = plt.subplots(figsize=(12,12))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_2)
plt.show()
```
